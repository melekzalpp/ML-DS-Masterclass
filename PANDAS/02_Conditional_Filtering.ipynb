## Multiple Conditions
Recall the steps:

* Get the conditions
* Wrap each condition in parenthesis
* Use the | or & operator, depending if you want an
    * OR | (either condition is True)
    * AND & (both conditions must be True)
* You can also use the ~ operator as a NOT operation

```python
df[(df["total_bill"] > 30) &  (df["sex"] == "Male")]
```

```python
df[(df["total_bill"] > 30) |  (df["sex"] == "Male")]
```

```python
df[(df["day"] == "Sun") | (df["day"] == "Sat") | (df["day"] == "Fri") ]
# this sort of thing to get really annoying if we're always using the same column.
# same column => "day"
# isin() method is quicker for that.
```

## Conditional Operator isin()
We can use .isin() operator to filter by a list of options.

```python
options = ["Sat", "Sun"]
```

```python
df["day"].isin(options)
# it is going to return true or false if the value for whatever particular road
# its analyzing in the day column is present in this list of options
```

```python
df[df["day"].isin(options)]
```

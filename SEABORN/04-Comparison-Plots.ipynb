# Comparison Plots with pairplot() and jointplot()

- Comparison plots are essentially 2-dimesnional versions of the plots we've learned about so far.

- The two main plots types discussed here: jointplot() and pairplot().

## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

- **jointplot()**

- We can map histograms to each feature of scatterplot to clarify the distributions within each feature.

- We can also adjust the scatterplot to be a hex plot or a 2D KDE plot.

## The Data

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DS&ML Masterclass/05-Seaborn/StudentsPerformance.csv")
```

```python
df.head()
```

## jointplot

```python
sns.jointplot(x='math score',y='reading score',data=df)
# it is a scatterplot of two continupuse features that adds in histograms.
```

```python
# Histograms with hexagons
# Hexagons are dark the more points fall into their area.
sns.jointplot(x='math score',y='reading score',data=df,kind='hex')
```

```python
# 2D KDE plot show shaded distribution between both KDEs
sns.jointplot(x='math score',y='reading score',data=df,kind='kde')
```

## pairplot

- The pairplot() is a quick way to compare all numerical columns in a DataFrame.

- It automatically creates a histogram for each column and a scatterplot comparison between all possible combinations of columns.

```python
sns.pairplot(df)
```

Warning!

- pairplot() can be CPU and RAM instensive for large DataFrames with many columns.

- It is a good idea to first filter down to only the columns you are insterested in.

```python
sns.pairplot(df,hue='gender',palette='viridis')
```

```python
sns.pairplot(df,hue='gender',palette='viridis',corner=True)
```

```python
sns.pairplot(df,hue='gender',palette='viridis',diag_kind='hist')
```
